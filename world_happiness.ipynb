```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
happiness_2015 = pd.read_csv('C:/datasets/happiness_ranking/2015.csv')

happiness_2015.columns = ['Country', 'Region', 'Happiness_rank', 'Happiness_score', 'Standart_error', 'Economy', 'Family',
              'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia_residual']

new_columns = happiness_2015.columns.values
for i in range(2, len(new_columns)):
    new_columns[i] += '_2015'
    
happiness_2015.columns = new_columns
happiness_2015.head()
```

```python
sns.set()
fig, axes = plt.subplots(2,3, figsize=(15,10))
fig.suptitle('How Happiness Score changes depend on 6 different factors - in 2015', size=17)

axes = axes.flatten()

for i in range(6):
    factor_name = happiness_2015.columns[i+5]
    #axes[i].set_title(factor_name)
    sns.regplot(ax=axes[i], x=factor_name, y='Happiness_score_2015', data=happiness_2015, scatter_kws={'alpha':0.3}, color='navy')
```

```python
g = sns.catplot(x='Region', y='Happiness_score_2015', data=happiness_2015, aspect=3, s=7, palette='mako')
g.fig.suptitle('Happiness Score in different Regions', size=17, position=(0.5,1.1))  

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)
```

```python
g = sns.catplot(x='Region', y='Happiness_score_2015', kind='boxen', data=happiness_2015, aspect=3, palette='mako')
g.fig.suptitle('Happiness Score in different Regions', size=17, position=(0.5,1.1))  

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)
```

```python
happiness_2017 = pd.read_csv('C:/datasets/happiness_ranking/2017.csv')

happiness_2017.columns = ['Country', 'Happiness_rank', 'Happiness_score', 'Whisker_high', 'Whisker_low', 'Economy', 'Family',
              'Health', 'Freedom', 'Generosity', 'Trust', 'Dystopia_residual']

new_columns = happiness_2017.columns.values
for i in range(1, len(new_columns)):
    new_columns[i] += '_2017'
    
happiness_2017.columns = new_columns
happiness_2017.head()
```

```python
happiness_2019 = pd.read_csv('C:/datasets/happiness_ranking/2019.csv')

happiness_2019.columns = ['Happiness_rank', 'Country', 'Happiness_score', 'Economy', 'Family', 'Health', 'Freedom',
                          'Generosity', 'Trust']

new_columns = happiness_2019.columns.values
for i in range(len(new_columns)):
    if i==1:
        continue
    new_columns[i] += '_2019'
    
happiness_2019.columns = new_columns
happiness_2019.head()
```

```python
#joining tables for 2015, 2017 and 2019 years accordong to countries
df = pd.merge(happiness_2015, happiness_2017, on='Country')
df = pd.merge(df, happiness_2019, on='Country')
df.head()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,10))
fig.suptitle('Trends of 6 different factors - in 2015, 2017 and 2019', size=17)

axes = axes.flatten()

factor_list = ['Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity']

for i, factor in enumerate(factor_list):
    sns.regplot(ax=axes[i], x='Happiness_rank_2015', y=factor + '_2015', data=df, scatter_kws={'alpha':0.1}, color='yellow')
    sns.regplot(ax=axes[i], x='Happiness_rank_2017', y=factor + '_2017', data=df, scatter_kws={'alpha':0.1}, color='grey')
    sns.regplot(ax=axes[i], x='Happiness_rank_2019', y=factor + '_2019', data=df, scatter_kws={'alpha':0.1}, color='navy')
    axes[i].set_xlabel('Happiness rank')
    axes[i].set_ylabel(factor)
    axes[i].legend(['2015', '2017', '2019'])
```

```python
fig, axes = plt.subplots(figsize=(7,7))
fig.suptitle('Happiness Score trends in 2015, 2017 and 2019', size=17)
sns.lineplot(x="Happiness_rank_2015", y="Happiness_score_2015", data=df)
sns.lineplot(x="Happiness_rank_2017", y="Happiness_score_2017", data=df)
sns.lineplot(x="Happiness_rank_2019", y="Happiness_score_2019", data=df)
axes.legend(['2015','2017','2019'])
axes.set_xlabel('Happiness rank')
axes.set_ylabel('Happiness score')
```

```python
#let's try to investigate if any country experienced a sighnificant increase or decrease in Happiness Score..
#We'll make 2 new columns where we'll calculate the difference of Happiness Score for each country in 2017-2015 and 2019-2017 
#accordingly.

df['H_Sc_diff_17_15'] = df['Happiness_score_2017'] - df['Happiness_score_2015']
df['H_Sc_diff_19_17'] = df['Happiness_score_2019'] - df['Happiness_score_2017']

#and will find the maximum value

abs_column1 = abs(df['H_Sc_diff_17_15'])
abs_column2 = abs(df['H_Sc_diff_19_17'])
ind_max1 = abs_column1.idxmax()
ind_max2 = abs_column2.idxmax()

#plotting the country with max value of difference in 2015-2017
value2015 = df['Happiness_score_2015'][ind_max1]
value2017 = df['Happiness_score_2017'][ind_max1]
data = {'year_2015': value2015, 'year_2017': value2017}
df_max_country_15_17 = pd.DataFrame.from_dict(data, orient='index', columns=['Happiness score'])
fig, axes = plt.subplots() 
sns.lineplot(ax=axes, data=df_max_country_15_17, markers=True).set_title("""'The country with the biggest 
        difference of Happiness Score between 2015 and 2017 years -- """ + df['Country'][ind_max1], size=17)
axes.set_ylim(0,8)
```

```python
#plotting the country with max value of difference in 2017-2019
value2017 = df['Happiness_score_2017'][ind_max2]
value2019 = df['Happiness_score_2019'][ind_max2]
data = {'year_2017': value2017, 'year_2019': value2019}
df_max_country_17_19 = pd.DataFrame.from_dict(data, orient='index', columns=['Happiness score'])
fig, axes = plt.subplots() 
sns.lineplot(ax=axes, data=df_max_country_17_19, markers=True).set_title("""'The country with the biggest 
        difference of Happiness Score between 2017 and 2019 years -- """ + df['Country'][ind_max2], size=17)
axes.set_ylim(0,8)
```
